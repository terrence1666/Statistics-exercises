# Exercise 3. Data Management with Python

## 3.1. Import libraries

```python
import numpy as np
import pandas as pd
import scipy.stats as st
import datetime
import matplotlib.pyplot as plt
```

## 3.2. Read data 

```python
data = pd.read_csv("elbedata.txt", delimiter = '\t')
print (data)
```

```python
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['doy'] = data['date'].dt.dayofyear
data['week'] = data['date'].dt.week
print(data)
```

## 3.3. Subsetting

```python
data2 = data.set_index('date')
year2002 = data2.loc['2002', 'discharge']
jan2002 = data2.loc['2002','discharge']
print (data2,year2002,jan2002)
year2002.mean()
```

## 3.4 Plotting

```python
plt.plot(data['doy'],data['discharge'],'.',markersize = 2,label = 'Discharge')
plt.xlabel('Day of year')
plt.ylabel('Discharge ($m^3$/d)')
plt.legend()
plt.show()
index = ['year','month','day','doy','week']
for i in index:
    data.boxplot(by = i, column = ['discharge'])
```

```python
print(data['discharge'].describe())
data['discharge'].hist()
plt.xlabel('Discharge')
plt.ylabel('Frequency')
```
